```python
# 1. Install the Modern Packages
!pip install -q -U trulens-eval trulens-providers-google trulens-apps-llamaindex llama-index-llms-google-genai llama-index-embeddings-google-genai nest_asyncio

# 2. Async Patch
import nest_asyncio
nest_asyncio.apply()

import os
import numpy as np
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings
from llama_index.llms.google_genai import GoogleGenAI
from llama_index.embeddings.google_genai import GoogleGenAIEmbedding

# 3. Modern TruLens Imports
from trulens.core import TruSession, Metric, Selector
from trulens.apps.llamaindex import TruLlama
from trulens.providers.google import Google
from trulens.dashboard import run_dashboard

# 4. Connect Keys
os.environ["GOOGLE_API_KEY"] = "AIzaSyDij6eHWZRTk24L0iqoVR9YqRAIIOXUz9A"

# 5. LlamaIndex Configuration
Settings.llm = GoogleGenAI(model="models/gemini-3-flash-preview")
Settings.embed_model = GoogleGenAIEmbedding(model_name="models/gemini-embedding-2-preview")

print("📂 Loading IEEE GUC data...")
documents = SimpleDirectoryReader(input_dir="./mydata").load_data()
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine()

# 6. Initialize Modern TruLens
print("⚖️ Setting up the Modern RAG Triad...")
session = TruSession()
session.reset_database()
provider = Google()

# ==========================================
# 🛠️ THE FIX: WRAPPER FUNCTIONS
# We define the exact variable names TruLens is looking for to bypass the signature bug.
# ==========================================
def custom_answer_relevance(prompt: str, response: str):
    return provider.relevance_with_cot_reasons(prompt, response)

def custom_context_relevance(question: str, context: str):
    return provider.qs_relevance_with_cot_reasons(question, context)

def custom_groundedness(source: str, statement: str):
    return provider.groundedness_measure_with_cot_reasons(source, statement)
# ==========================================

# 7. Define Rubrics (Using our new wrappers!)
f_answer_relevance = Metric(
    name="Answer Relevance",
    implementation=custom_answer_relevance,
    selectors={
        "prompt": Selector.select_record_input(),
        "response": Selector.select_record_output(),
    }
)

f_context_relevance = Metric(
    name="Context Relevance",
    implementation=custom_context_relevance,
    selectors={
        "question": Selector.select_record_input(),
        "context": Selector.select_context(collect_list=False),
    },
    agg=np.mean
)

f_groundedness = Metric(
    name="Groundedness",
    implementation=custom_groundedness,
    selectors={
        "source": Selector.select_context(collect_list=True),
        "statement": Selector.select_record_output(),
    }
)

# 8. Attach the Grader & Test
tru_recorder = TruLlama(
    query_engine,
    app_name='IEEE_GUC_Assistant_vModern',
    feedbacks=[f_answer_relevance, f_context_relevance, f_groundedness]
)

print("🤖 Taking the exam... (Waiting 10-15 seconds for grading)")
with tru_recorder as recording:
    response = query_engine.query("What specific events does IEEE GUC have planned for 2026, and how do they relate to media?")

print(f"\n✅ Test Complete! AI Answer: {response}")

# 9. Launch the Dashboard
run_dashboard(session)
```

```python
import time
from llama_index.core import Settings

print("🛡️ Engaging 'God Mode': Bypassing TruLens due to Google API schema bugs...\n")

# The AI engine we already know works perfectly
llm = Settings.llm

# 1. Ask the Question
question = "What specific events does IEEE GUC have planned for 2026, and how do they relate to media?"
response = query_engine.query(question)
answer = str(response)

# Extract the exact paragraphs the AI read from your document
context = "\n".join([node.node.text for node in response.source_nodes])

print(f"🤖 AI Answer:\n{answer}\n")
print("⏳ Starting Direct LLM Grading...\n")

# 2. Custom Grader Function
def direct_grade(prompt_text):
    try:
        # We use the LLM directly, skipping TruLens's broken formatting
        res = llm.complete(prompt_text)
        return res.text.strip()
    except Exception as e:
        return f"Grading failed: {e}"

# 3. Grade Answer Relevance
time.sleep(2)
print("📊 Grading Answer Relevance (Did it answer the question?)...")
prompt_ans = f"Evaluate this Answer based on the Question.\nQuestion: {question}\nAnswer: {answer}\nProvide a Score from 0.0 to 1.0. Format exactly as 'Score: X.X - Reason: ...'"
print(f"   ➤ {direct_grade(prompt_ans)}\n")

# 4. Grade Context Relevance
time.sleep(2)
print("📊 Grading Context Relevance (Did LlamaIndex find the right text?)...")
prompt_ctx = f"Evaluate if this Context contains the answer to the Question.\nQuestion: {question}\nContext: {context}\nProvide a Score from 0.0 to 1.0. Format exactly as 'Score: X.X - Reason: ...'"
print(f"   ➤ {direct_grade(prompt_ctx)}\n")

# 5. Grade Groundedness
time.sleep(2)
print("📊 Grading Groundedness (Is the AI hallucinating?)...")
prompt_grnd = f"Evaluate if this Answer is 100% supported by the Context.\nContext: {context}\nAnswer: {answer}\nProvide a Score from 0.0 to 1.0. Format exactly as 'Score: X.X - Reason: ...'"
print(f"   ➤ {direct_grade(prompt_grnd)}\n")

print("✅ ADVANCED RAG PIPELINE OFFICIALLY COMPLETE!")
```
